# Importing Libraries


```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

# Load the dataset


```python
dataset = pd.read_csv("concrete_data.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

# Normalize the input features


```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

# Define Regression Model Function


```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation="relu", input_shape=(X_normalized.shape[1],)))
    model.add(Dense(1))
    model.compile(optimizer="adam", loss="mean_squared_error")
    return model
```

# Loop 50 Times


```python
mse_list_normalized = []

for _ in range(50):
    # Split normalized data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3)

    # Create and train model
    model = regression_model()
    model.fit(X_train, y_train, epochs=50, verbose=str(0))

    # Evaluate model
    y_pred = model.predict(X_test)
    mse_normalized = mean_squared_error(y_test, y_pred)
    mse_list_normalized.append(mse_normalized)
```

# Calculate mean and standard deviation of mean squared errors


```python
mean_mse_normalized = np.mean(mse_list_normalized)
std_mse_normalized = np.std(mse_list_normalized)

print("Mean of Mean Squared Errors (Normalized):", mean_mse_normalized)
print("Standard Deviation of Mean Squared Errors (Normalized):", std_mse_normalized)
```
